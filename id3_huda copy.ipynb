```python
import pandas as pd
df = pd.read_csv('diabetes_prediction_dataset.csv',encoding='latin1')
df
```

```python
x= df.iloc[:,0:-1]
y= df.iloc[:,[-1]]

from sklearn.preprocessing import LabelEncoder as LE
encode = LE()
x['gender'] = encode.fit_transform(x['gender'])
x['smoking_history'] = encode.fit_transform(x['smoking_history'])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=99)
```

```python
import pickle
model = DecisionTreeClassifier(criterion="entropy")
model = model.fit(x_train,y_train)

with open('diabetes_decision_tree_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

y_pred = model.predict( x_test)
y_pred
```

```python
from sklearn import metrics
print("accuracy : ", metrics.accuracy_score(y_test,y_pred)*100)
# databaru = pd.read_csv('diabetes_dataset.csv')
# new_data = databaru.iloc[0:27,:-1]
```
